```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
yasListesi = [10, 20, 30, 30, 30, 40, 50, 60, 70, 75]
kiloListesi = [20, 60, 80, 85, 86, 87, 70, 90, 95, 90]

numpyYasListesi = np.array(yasListesi)
numpyKiloListesi = np.array(kiloListesi)
```

```python
numpyYasListesi
```

```python
numpyKiloListesi
```

```python
plt.plot(numpyYasListesi, numpyKiloListesi, "purple")
plt.xlabel("yaş")
plt.ylabel("kilo")
plt.title("kilonun yaşa göre değişim grafiği")
plt.show()
```

```python
numpyDizisi1 = np.linspace(0,10,20)
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 ** 3
numpyDizisi2
```

```python
plt.plot(numpyDizisi1, numpyDizisi2, "b")  
plt.show()
```

```python
plt.subplot(1,2,1) # 1 sıra olacak, 2 kolon olacak, şuan 1. grafik çiziliyor
plt.plot(numpyDizisi1,numpyDizisi2, "r--")
plt.subplot(1,2,2) # 1 sıra olacak, 2 kolon olacak, şuan 2. grafik çiziliyor
plt.plot(numpyDizisi2,numpyDizisi1, "g*-")
plt.show()
```

```python
figure1 = plt.figure() #boş bir kanvas, eksenler yok

figureAxes = figure1.add_axes([0.4,0.4, 0.6,0.6])
# 0.4 demek: Figürün solundan %40 uzaklıkta başla.
# 0.4 demek: Figürün altından %40 yüksekte başla.
# 0.6 demek: Figürün yatay boyutunun %60'ı kadar genişlik.
# 0.6 demek: Figürün dikey boyutunun %60'ı kadar yükseklik.
figureAxes.plot(numpyDizisi1, numpyDizisi2, "b")
figureAxes.set_xlabel("X ekseni")
figureAxes.set_ylabel("Y ekseni")
figureAxes.set_title("Başlık")
plt.show()
```

## **iç içe grafikler**

```python
figure2 = plt.figure()

eksen1 = figure2.add_axes([0.1,0.1,0.77,0.77])
eksen2 = figure2.add_axes([0.3,0.5,0.3,0.3])

eksen1.plot(numpyDizisi1,numpyDizisi2,"g")
eksen1.set_xlabel("x ekseni")
eksen1.set_ylabel("y ekseni")
eksen1.set_title("ana grafik başlık")

eksen2.plot(numpyDizisi2,numpyDizisi1,"b")
eksen2.set_xlabel("x ekseni")
eksen2.set_ylabel("y ekseni")
eksen2.set_title("küçük grafik başlık")

plt.show()
```

## **subplots**

```python
plt.subplots(figsize=(2,2))
plt.show()
```

```python
(figure1, benimEksenler) = plt.subplots(figsize=(1,1))
plt.show()
```

```python
figure1
```

```python
(figure1, benimEksenler) = plt.subplots(nrows=1, ncols=2)
type(benimEksenler)

for eksen in benimEksenler:
    eksen.plot(numpyDizisi1, numpyDizisi2, "g")
    eksen.set_xlabel("x ekseni")

plt.tight_layout()
plt.show()
```

```python
(figure1, benimEksenler) = plt.subplots(nrows=1, ncols=2)

benimEksenler[0].plot(numpyDizisi1, numpyDizisi2, "g")
benimEksenler[0].set_title("Grafik 1")
benimEksenler[0].set_xlabel("x ekseni")

benimEksenler[1].plot(numpyDizisi2, numpyDizisi1, "b")
benimEksenler[1].set_title("Grafik 2")
benimEksenler[1].set_xlabel("x ekseni")

plt.tight_layout()
plt.show()
```

```python
veriler = [
    (numpyDizisi1, numpyDizisi2, "g", "Grafik 1"),
    (numpyDizisi2, numpyDizisi1, "b", "Grafik 2")
]

(figure1, benimEksenler) = plt.subplots(nrows=1, ncols=2, figsize = (5, 2))

for eksen, (x_verisi, y_verisi, renk, baslik) in zip(benimEksenler, veriler):
    eksen.plot(x_verisi, y_verisi, renk)
    eksen.set_title(baslik)
    eksen.set_xlabel("x ekseni")

plt.tight_layout()
plt.show()
```

>**`görsel geliştirmeler`**

```python
yeniFigur = plt.figure(dpi = 150)
# 'dpi' grafik kalitesi

yeniEksen = yeniFigur.add_axes([0.1,0.1,0.5,0.5])

yeniEksen.plot(numpyDizisi1, numpyDizisi1**2, label = "numpyDizisi ** 2")
yeniEksen.plot(numpyDizisi1, numpyDizisi1**3, label = "numpyDizisi ** 3")
yeniEksen.legend(loc=1)

plt.show()
```

>**`görsel geliştirmeler ileri seviye`**

```python
numpyDizisi2 = numpyDizisi1 ** 2

(figur, eksen) = plt.subplots(figsize = (5,3))

eksen.plot(numpyDizisi1, numpyDizisi2, color = "#3A95A8")
eksen.plot(numpyDizisi2, numpyDizisi1, color = "r", alpha = 0.4)  # "alpha" saydamlık katar
plt.show()
```

```python
(fig, eks) = plt.subplots(figsize = (5,3), dpi=150)

eks.plot(numpyDizisi1, numpyDizisi1 + 2, color = "purple", linewidth = 6, linestyle = "dashdot")
eks.plot(numpyDizisi1, numpyDizisi1 + 3, color = "#c0eb34", linestyle = "--")
eks.plot(numpyDizisi1, numpyDizisi1 + 4, color = "#34eb6b", linestyle = "dotted")
eks.plot(numpyDizisi1, numpyDizisi1 + 5, color = "#3455eb", linestyle = ":", marker = "o", markersize = 4,markerfacecolor = "red")
eks.plot(numpyDizisi1, numpyDizisi1 + 6, color = "#3455eb", linestyle = "solid", marker = "+", markersize = 8)

plt.show()
```
